```python
import cv2

img = cv2.imread('img_2.jfif')
old_img = cv2.resize(img, (600, 500))  # Set a reasonable width

help(cv2.ellipse)
help(cv2.rectangle)
help(cv2.circle)
new_img = cv2.rectangle(old_img, pt1=(140,0), pt2=(380,300),
                       color=(0,255,0), thickness=4, lineType=4,
                       shift=0)
new_img = cv2.ellipse(old_img, center=(250,180), axes=(100,120),
                     angle=30, startAngle=0, endAngle=360,
                     color=(0,0,255), thickness=4, lineType=1, shift=0)

new_img = cv2.line(old_img, pt1=(170,125) , pt2=(290,115) ,
                   color=(0,255,0), thickness=4, lineType=1, shift=0)

new_img = cv2.circle(old_img, center=(240, 180), radius=120, color=(255, 0, 0),
                     thickness=4, lineType=cv2.LINE_AA, shift=0)
cv2.imshow('my window', new_img)  # Show resized image
cv2.waitKey(0)
cv2.destroyAllWindows()
```
